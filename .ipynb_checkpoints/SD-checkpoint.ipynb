```python
from pathlib import Path

import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, set_seed

# Set Seed
set_seed(42)

# Check if GPU is available
device = "cuda" if torch.cuda.is_available() else "cpu"
print(device)
```

```python
cache_path = str((Path(".hf_cache") / "hub").resolve()) # project-local Hugging Face hub cache

small_tokenizer = AutoTokenizer.from_pretrained("distilbert/distilgpt2", cache_dir=cache_path, local_files_only=True)
small_model = AutoModelForCausalLM.from_pretrained("distilbert/distilgpt2", cache_dir=cache_path, local_files_only=True).to(device)
large_tokenizer = AutoTokenizer.from_pretrained("openai-community/gpt2-large",cache_dir=cache_path,local_files_only=True)
large_model = AutoModelForCausalLM.from_pretrained("openai-community/gpt2-large",cache_dir=cache_path, local_files_only=True).to(device)
# large_tokenizer = AutoTokenizer.from_pretrained("openai-community/gpt2-large", cache_dir=cache_path, local_files_only=True)
# large_model = AutoModelForCausalLM.from_pretrained("openai-community/gpt2-large", cache_dir=cache_path, local_files_only=True).to(device)
```

```python
def normal_inference(model, tokenizer, prompt, max_new_tokens=30):
    inputs = tokenizer(prompt, return_tensors='pt').to(device)
    outputs = model.generate(inputs['input_ids'], max_new_tokens=max_new_tokens, attention_mask=inputs["attention_mask"],  
                             pad_token_id=tokenizer.eos_token_id, return_dict_in_generate=True, output_scores=True,)
                                # do_sample=True,        # enable sampling
                                # temperature=0.8,       # control randomness
                                # top_p=0.9)             # nucleus sampling
    text = tokenizer.decode(outputs.sequences[0], skip_special_tokens=True)
    input_len = inputs["input_ids"].shape[1]
    new_tokens = outputs.sequences[0][input_len:]

    distributions = []
    token_probs = []

    for token_id, score in zip(new_tokens, outputs.scores):
        prob_dist = torch.softmax(score[0], dim=-1)
        prob = prob_dist[token_id].item()

        distributions.append(prob_dist)
        token_probs.append((tokenizer.decode([token_id]), prob))

    return text, new_tokens, token_probs, distributions
```

```python
# =====================================================
# Algorithm 1: SpeculativeDecodingStep
# Inputs: target model Mp, draft model Mq, and prefix
# Here:
#   Mp = large_model
#   Mq = small_model
#   prefix = prompt / generated
# =====================================================

prompt = "Once upon a time, there was a little girl named Alice who lived in a small village."
max_new_tokens = 50
gamma = 4

small_model.eval()
large_model.eval()

inputs = tokenizer(prompt, return_tensors="pt").to(device)
input_ids = inputs["input_ids"]

# prefix
generated = input_ids.clone()

prompt_len = input_ids.shape[1]
target_len = prompt_len + max_new_tokens

accepted_tokens = 0
total_draft_tokens = 0


# Repeat speculative decoding steps until enough tokens
# are generated
while generated.shape[1] < target_len:
    # Step 1:
    # Sample gamma guesses x_1, ..., x_gamma from Mq
    # autoregressively.
    #
    # In the paper:
    #   q_i(x) <- Mq(prefix + [x_1, ..., x_{i-1}])
    #   x_i ~ q_i(x)
    #
    # In code:
    #   small_model.generate(...) generates gamma draft tokens.
    #   draft_outputs.scores stores q_1, ..., q_gamma.
    old_len = generated.shape[1]
    draft_outputs = small_model.generate(
        input_ids=generated,
        max_new_tokens=gamma,
        pad_token_id=tokenizer.eos_token_id,
        return_dict_in_generate=True,
        output_scores=True,
        # Greedy decoding by default.
        # For sampling-based SD, you may enable:
        # do_sample=True,
        # temperature=0.8,
        # top_p=0.9,
        )
    # Full sequence = prefix + draft tokens
    draft_sequence = draft_outputs.sequences[0]
    # Only newly generated draft tokens:
    # [x_1, ..., x_gamma]
    draft_tokens = draft_sequence[old_len:]
    if len(draft_tokens) == 0:
        break
    total_draft_tokens += len(draft_tokens)
    # Convert draft-model logits into probability distributions.
    # q_distributions[i] is q_{i+1}(x)
    q_distributions = [
        torch.softmax(score[0], dim=-1)
        for score in draft_outputs.scores
    ]
    print("\nDraft tokens from Mq:")
    for tok in draft_tokens:
        print(repr(tokenizer.decode([tok.item()])), end=" | ")
    print()
    # Step 2:
    # Run Mp in parallel.
    #
    # In the paper:
    #   p_1(x), ..., p_{gamma+1}(x) <-
    #       Mp(prefix), ..., Mp(prefix + [x_1, ..., x_gamma])
    #
    # Instead of calling Mp gamma+1 times, we feed:
    #   candidate_ids = prefix + [x_1, ..., x_gamma]
    #
    # One forward pass gives all target distributions:
    #   p_1, ..., p_{gamma+1}
    # =====================================================

    candidate_ids = torch.cat(
        [generated, draft_tokens.unsqueeze(0)],
        dim=1
    )

    with torch.no_grad():
        target_outputs = large_model(input_ids=candidate_ids)


    # =====================================================
    # Step 3:
    # Determine the number of accepted guesses n.
    # In the paper:
    #   r_i ~ U(0, 1)
    #   accept x_i if r_i <= p_i(x_i) / q_i(x_i)
    # If a token is rejected, stop checking later draft tokens.
    # The number of accepted tokens is n.
    accepted_list = []
    rejected = False
    reject_index = None
    for i, token_id in enumerate(draft_tokens):
        # q_i(x_i): draft model probability for token x_i
        q_dist = q_distributions[i]
        q_prob = q_dist[token_id].item()
        # p_i(x_i): target model probability for the same token x_i
        # Important indexing:
        # token x_i is located at position old_len + i.
        # The probability of this token is predicted by logits
        # at position old_len + i - 1.
        target_logits = target_outputs.logits[0, old_len + i - 1, :]
        p_dist = torch.softmax(target_logits, dim=-1)
        p_prob = p_dist[token_id].item()
        # Acceptance probability:min(1, p_i(x_i) / q_i(x_i))
        if q_prob > 0:
            accept_prob = min(1.0, p_prob / q_prob)
        else:
            accept_prob = 1.0
        # r_i ~ U(0, 1)
        random_number = torch.rand(1).item()
        token_text = tokenizer.decode([token_id.item()])
        print("\nChecking token:", repr(token_text))
        print("q_i(x_i) from Mq:", q_prob)
        print("p_i(x_i) from Mp:", p_prob)
        print("acceptance probability:", accept_prob)
        print("random number r_i:", random_number)
        # If r_i <= p_i(x_i) / q_i(x_i), accept this token.
        if random_number <= accept_prob:
            print("Decision: ACCEPT")
            accepted_list.append(token_id.view(1, 1))
            accepted_tokens += 1
        # Otherwise, reject this token and stop.
        else:
            print("Decision: REJECT")
            rejected = True
            reject_index = i
            break
    # Step 4:
    # Return n tokens from Mq.
    #
    # In the paper:
    #   return prefix + [x_1, ..., x_n, t]
    #
    # Here we first add the accepted draft tokens:
    #   [x_1, ..., x_n]
    # =====================================================

    if len(accepted_list) > 0:
        accepted_tensor = torch.cat(accepted_list, dim=1)
        generated = torch.cat([generated, accepted_tensor], dim=1)

    if generated.shape[1] >= target_len:
        break


    # =====================================================
    # Step 5A:
    # Adjust the distribution from Mp if needed.
    #
    # If n < gamma, meaning one draft token was rejected:
    #
    # In the paper:
    #   p'(x) <- norm(max(0, p_{n+1}(x) - q_{n+1}(x)))
    #   t ~ p'(x)
    #
    # This removes the probability mass already over-proposed
    # by the draft model.
    # =====================================================

    if rejected:
        q_dist = q_distributions[reject_index]

        target_logits = target_outputs.logits[0, old_len + reject_index - 1, :]
        p_dist = torch.softmax(target_logits, dim=-1)

        # p'(x) = norm(max(0, p(x) - q(x)))
        adjusted_dist = torch.clamp(p_dist - q_dist, min=0)

        if adjusted_dist.sum().item() == 0:
            adjusted_dist = p_dist
        else:
            adjusted_dist = adjusted_dist / adjusted_dist.sum()

        # t ~ p'(x)
        next_token = torch.multinomial(
            adjusted_dist,
            num_samples=1
        ).view(1, 1)

        print("\nRejected case:")
        print("Replacement token t from adjusted distribution p'(x):",
              repr(tokenizer.decode([next_token.item()])))

        generated = torch.cat([generated, next_token], dim=1)


    # =====================================================
    # Step 5B:
    # If all gamma guesses are accepted, no adjustment is needed.
    # In the paper:
    #   p'(x) <- p_{gamma+1}(x)
    #   t ~ p'(x)
    # This means we sample one extra token from the target model.
    else:
        # p_{gamma+1}(x): target distribution after all draft tokens
        next_logits = target_outputs.logits[0, -1, :]
        next_dist = torch.softmax(next_logits, dim=-1)

        # t ~ p_{gamma+1}(x)
        next_token = torch.multinomial(
            next_dist,
            num_samples=1
        ).view(1, 1)

        print("\nAll draft tokens accepted:")
        print("Extra token t from Mp:",
              repr(tokenizer.decode([next_token.item()])))

        generated = torch.cat([generated, next_token], dim=1)
    # Optional:
    # Stop if EOS token is generated.
    # =====================================================
    if tokenizer.eos_token_id is not None:
        if generated[0, -1].item() == tokenizer.eos_token_id:
            break
# Final result:
# return prefix + [x_1, ..., x_n, t]
# repeated until max_new_tokens is reached.
# =====================================================
final_ids = generated[0, :target_len]
final_text = tokenizer.decode(final_ids, skip_special_tokens=True)
# Remove newlines and repeated spaces
final_text = " ".join(final_text.split())
acceptance_rate = accepted_tokens / total_draft_tokens if total_draft_tokens > 0 else 0
print("\n==============================")
print("Final text:")
print(final_text)
print("\nAccepted draft tokens:", accepted_tokens)
print("Total draft tokens:", total_draft_tokens)
print("Acceptance rate:", acceptance_rate)
```

```python
# ==================================================
# CUDA benchmark with greedy verification
# ==================================================

device = "cuda"
print(device)

prompt = "Once upon a time, there was a little girl named Alice who lived in a small village."
max_new_tokens = 512
gammas = [1, 2, 4, 8]

# target "gpt2-large" 774M
# draft "distilgpt2" 82M

tokenizer = small_tokenizer
tokenizer.pad_token = tokenizer.eos_token

target_model = large_model.to(device)

draft_model = small_model.to(device)

target_model.eval()
draft_model.eval()

inputs = tokenizer(prompt, return_tensors="pt").to(device)
input_ids = inputs["input_ids"]

prompt_len = input_ids.shape[1]
target_len = prompt_len + max_new_tokens


# ==================================================
# AR baseline
# ==================================================

torch.cuda.synchronize()
start_event = torch.cuda.Event(enable_timing=True)
end_event = torch.cuda.Event(enable_timing=True)
start_event.record()

with torch.no_grad():
    ar_outputs = target_model.generate(
        input_ids=input_ids,
        attention_mask=inputs["attention_mask"],
        max_new_tokens=max_new_tokens,
        pad_token_id=tokenizer.eos_token_id,
        return_dict_in_generate=True,
    )

end_event.record()
torch.cuda.synchronize()
ar_time = start_event.elapsed_time(end_event) / 1000
ar_ids = ar_outputs.sequences[:, :target_len]
ar_tokens = ar_ids.shape[1] - prompt_len
ar_speed = ar_tokens / ar_time

print(f"AR: {ar_time:.3f}s | {ar_speed:.3f} tok/s | tokens={ar_tokens}")


# ==================================================
# Speculative decoding
# ==================================================

for gamma in gammas:

    generated = input_ids.clone()

    accepted_tokens = 0
    total_draft_tokens = 0
    draft_past = None
    target_past = None
    draft_cache_len = 0
    target_cache_len = 0

    torch.cuda.synchronize()
    start_event = torch.cuda.Event(enable_timing=True)
    end_event = torch.cuda.Event(enable_timing=True)
    start_event.record()

    with torch.no_grad():

        while generated.shape[1] < target_len:

            # Step 1:
            # Generate gamma draft tokens from Mq.
            old_len = generated.shape[1]
            draft_gamma = min(gamma, target_len - old_len)

            draft_outputs = draft_model(
                input_ids=generated[:, draft_cache_len:],
                past_key_values=draft_past,
                use_cache=True
            )
            draft_past = draft_outputs.past_key_values

            draft_list = []

            for i in range(draft_gamma):
                next_token = torch.argmax(
                    draft_outputs.logits[:, -1, :],
                    dim=-1
                ).view(1, 1)
                draft_list.append(next_token)

                if i < draft_gamma - 1:
                    draft_outputs = draft_model(
                        input_ids=next_token,
                        past_key_values=draft_past,
                        use_cache=True
                    )
                    draft_past = draft_outputs.past_key_values

            draft_tokens = torch.cat(draft_list, dim=1)[0]

            if len(draft_tokens) == 0:
                break

            total_draft_tokens += len(draft_tokens)

            # Step 2:
            # Run Mp in parallel.
            candidate_ids = torch.cat(
                [generated[:, target_cache_len:], draft_tokens.unsqueeze(0)],
                dim=1
            )

            target_outputs = target_model(
                input_ids=candidate_ids,
                past_key_values=target_past,
                use_cache=True
            )
            target_past = target_outputs.past_key_values

            # Step 3:
            # Determine accepted draft tokens.
            accepted_list = []
            rejected = False
            reject_index = None

            for i, token_id in enumerate(draft_tokens):

                target_index = old_len + i - 1 - target_cache_len
                target_logits = target_outputs.logits[0, target_index, :]
                target_token = torch.argmax(target_logits)

                if token_id == target_token:
                    accepted_list.append(token_id.view(1, 1))
                    accepted_tokens += 1
                else:
                    rejected = True
                    reject_index = i
                    next_token = target_token.view(1, 1)
                    break

            # Step 4:
            # Add accepted draft tokens.
            if len(accepted_list) > 0:
                accepted_tensor = torch.cat(accepted_list, dim=1)
                generated = torch.cat([generated, accepted_tensor], dim=1)

            if generated.shape[1] >= target_len:
                break

            # Step 5A:
            # If one draft token is rejected, use the target-model token.
            if rejected:
                generated = torch.cat([generated, next_token], dim=1)

            # Step 5B:
            # If all draft tokens are accepted, add one extra token from Mp.
            else:

                next_logits = target_outputs.logits[0, -1, :]
                next_token = torch.argmax(next_logits).view(1, 1)

                generated = torch.cat([generated, next_token], dim=1)

                draft_outputs = draft_model(
                    input_ids=draft_tokens[-1].view(1, 1),
                    past_key_values=draft_past,
                    use_cache=True
                )
                draft_past = draft_outputs.past_key_values

            # Keep the cache through all confirmed tokens except the last one.
            draft_cache_len = generated.shape[1] - 1
            target_cache_len = generated.shape[1] - 1
            draft_past.crop(draft_cache_len)
            target_past.crop(target_cache_len)

            # Stop if EOS token is generated.
            if tokenizer.eos_token_id is not None:
                if generated[0, -1].item() == tokenizer.eos_token_id:
                    break

    end_event.record()
    torch.cuda.synchronize()
    sd_time = start_event.elapsed_time(end_event) / 1000

    final_ids = generated[0, :target_len]
    sd_tokens = len(final_ids) - prompt_len
    sd_speed = sd_tokens / sd_time
    speedup = ar_time / sd_time
    match = torch.equal(final_ids, ar_ids[0])

    acceptance_rate = accepted_tokens / total_draft_tokens if total_draft_tokens > 0 else 0

    print(
        f"SD gamma={gamma}: {sd_time:.3f}s | "
        f"{sd_speed:.3f} tok/s | "
        f"speedup={speedup:.3f}x | "
        f"accepted={accepted_tokens} | "
        f"draft={total_draft_tokens} | "
        f"acceptance_rate={acceptance_rate:.3f} | "
        f"match={match}"
    )
```

```python
# ==================================================
# Adaptive speculative decoding with random sampling
# ==================================================

from pathlib import Path

import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, set_seed

device = "cuda"
cache_path = str((Path(".hf_cache") / "hub").resolve())

prompt = "Once upon a time, there was a little girl named Alice who lived in a small village."
max_new_tokens = 512

# Strategy parameters
confidence_threshold = 0.05
margin_threshold = 0.02
entropy_threshold = 0.7

tokenizer = AutoTokenizer.from_pretrained(
    "distilbert/distilgpt2",
    cache_dir=cache_path,
    local_files_only=True
)
tokenizer.pad_token = tokenizer.eos_token

target_model = large_model.to(device)
draft_model = small_model.to(device)

target_model.eval()
draft_model.eval()

inputs = tokenizer(prompt, return_tensors="pt").to(device)
input_ids = inputs["input_ids"]

prompt_len = input_ids.shape[1]
target_len = prompt_len + max_new_tokens

fixed_gammas = [1, 2, 4, 8]
min_gamma = 1
max_gamma = 8


# ==================================================
# CUDA warm-up
# ==================================================

with torch.no_grad():
    for model in [target_model, draft_model]:
        model.generate(
            input_ids=input_ids,
            attention_mask=inputs["attention_mask"],
            max_new_tokens=8,
            pad_token_id=tokenizer.eos_token_id,
            do_sample=True,
            temperature=1.0,
            use_cache=True,
        )

torch.cuda.synchronize()


# ==================================================
# AR baseline
# ==================================================

set_seed(42)
torch.cuda.synchronize()
start_event = torch.cuda.Event(enable_timing=True)
end_event = torch.cuda.Event(enable_timing=True)
start_event.record()

with torch.no_grad():
    ar_outputs = target_model.generate(
        input_ids=input_ids,
        attention_mask=inputs["attention_mask"],
        max_new_tokens=max_new_tokens,
        pad_token_id=tokenizer.eos_token_id,
        do_sample=True,
        temperature=1.0,
        use_cache=True,
    )

end_event.record()
torch.cuda.synchronize()
ar_time = start_event.elapsed_time(end_event) / 1000
ar_tokens = ar_outputs.shape[1] - prompt_len
ar_speed = ar_tokens / ar_time

print(f"AR: {ar_time:.3f}s | {ar_speed:.3f} tok/s | tokens={ar_tokens}")


# ==================================================
# Speculative decoding
# ==================================================

def run_sd(strategy, fixed_gamma=4):

    generated = input_ids.clone()

    accepted_tokens = 0
    total_draft_tokens = 0
    draft_steps = 0
    current_gamma = fixed_gamma

    draft_past = None
    target_past = None
    draft_cache_len = 0
    target_cache_len = 0

    torch.cuda.synchronize()
    start_event = torch.cuda.Event(enable_timing=True)
    end_event = torch.cuda.Event(enable_timing=True)
    start_event.record()

    with torch.no_grad():

        while generated.shape[1] < target_len:

            old_len = generated.shape[1]

            if strategy in ["confidence", "margin", "entropy"]:
                gamma = max_gamma
            else:
                gamma = current_gamma
            gamma = min(gamma, target_len - old_len)

            # Step 1:
            # Sample draft tokens from Mq using KV cache.
            draft_outputs = draft_model(
                input_ids=generated[:, draft_cache_len:],
                past_key_values=draft_past,
                use_cache=True
            )
            draft_past = draft_outputs.past_key_values

            draft_list = []
            q_distributions = []

            for i in range(gamma):
                q_dist = torch.softmax(
                    draft_outputs.logits[:, -1, :],
                    dim=-1
                )
                next_token = torch.multinomial(q_dist, num_samples=1)

                draft_list.append(next_token)
                q_distributions.append(q_dist[0])

                confidence = q_dist[0, next_token.item()].item()
                is_eos = tokenizer.eos_token_id is not None and next_token.item() == tokenizer.eos_token_id

                if strategy == "confidence" and confidence < confidence_threshold:
                    break
                if strategy == "margin":
                    top_probs = torch.topk(q_dist[0], k=2).values
                    margin = (top_probs[0] - top_probs[1]).item()
                    if margin < margin_threshold:
                        break
                if strategy == "entropy":
                    entropy = -torch.sum(
                        q_dist[0] * torch.log(q_dist[0].clamp_min(1e-12))
                    ) / torch.log(q_dist.new_tensor(q_dist.shape[-1]))
                    if entropy.item() > entropy_threshold:
                        break

                if is_eos:
                    break

                if i < gamma - 1:
                    draft_outputs = draft_model(
                        input_ids=next_token,
                        past_key_values=draft_past,
                        use_cache=True
                    )
                    draft_past = draft_outputs.past_key_values

            draft_tokens = torch.cat(draft_list, dim=1)[0]
            total_draft_tokens += len(draft_tokens)
            draft_steps += 1

            # Step 2:
            # Verify all draft tokens with Mp in parallel.
            candidate_ids = torch.cat(
                [generated[:, target_cache_len:], draft_tokens.unsqueeze(0)],
                dim=1
            )

            target_outputs = target_model(
                input_ids=candidate_ids,
                past_key_values=target_past,
                use_cache=True
            )
            target_past = target_outputs.past_key_values

            # Step 3:
            # Accept with min(1, p(x) / q(x)).
            draft_count = len(draft_tokens)
            target_start = old_len - 1 - target_cache_len

            q_distributions = torch.stack(q_distributions)
            p_distributions = torch.softmax(
                target_outputs.logits[0, target_start:target_start + draft_count, :],
                dim=-1
            )

            token_indices = draft_tokens.unsqueeze(1)
            q_probs = q_distributions.gather(1, token_indices).squeeze(1)
            p_probs = p_distributions.gather(1, token_indices).squeeze(1)
            accept_probs = torch.clamp(
                p_probs / q_probs.clamp_min(1e-12),
                max=1.0
            )
            accepted = torch.rand_like(accept_probs) <= accept_probs

            positions = torch.arange(draft_count, device=device)
            reject_index = torch.where(
                accepted,
                draft_count,
                positions
            ).min().item()

            accepted_count = reject_index
            accepted_tokens += accepted_count
            rejected = reject_index < draft_count

            if rejected:
                q_dist = q_distributions[reject_index]
                p_dist = p_distributions[reject_index]
                adjusted_dist = torch.clamp(p_dist - q_dist, min=0)
                adjusted_sum = adjusted_dist.sum()
                adjusted_dist = torch.where(
                    adjusted_sum > 0,
                    adjusted_dist / adjusted_sum.clamp_min(1e-12),
                    p_dist
                )
                next_token = torch.multinomial(
                    adjusted_dist,
                    num_samples=1
                ).view(1, 1)

            # Step 4:
            # Add accepted draft tokens.
            if accepted_count > 0:
                accepted_tensor = draft_tokens[:accepted_count].unsqueeze(0)
                generated = torch.cat([generated, accepted_tensor], dim=1)

            if generated.shape[1] >= target_len:
                break

            if tokenizer.eos_token_id is not None:
                if generated[0, -1].item() == tokenizer.eos_token_id:
                    break

            # Step 5A:
            # Sample from norm(max(0, p - q)) after rejection.
            if rejected:
                generated = torch.cat([generated, next_token], dim=1)

            # Step 5B:
            # Sample one extra token from Mp if all guesses are accepted.
            else:
                next_logits = target_outputs.logits[0, -1, :]
                next_dist = torch.softmax(next_logits, dim=-1)
                next_token = torch.multinomial(
                    next_dist,
                    num_samples=1
                ).view(1, 1)

                generated = torch.cat([generated, next_token], dim=1)

                draft_outputs = draft_model(
                    input_ids=draft_tokens[-1].view(1, 1),
                    past_key_values=draft_past,
                    use_cache=True
                )
                draft_past = draft_outputs.past_key_values

            # Update heuristic gamma.
            if strategy == "heuristic":
                if rejected:
                    current_gamma = max(min_gamma, current_gamma - 1)
                else:
                    current_gamma = min(max_gamma, current_gamma + 2)

            # Keep the cache through all confirmed tokens except the last one.
            draft_cache_len = generated.shape[1] - 1
            target_cache_len = generated.shape[1] - 1
            draft_past.crop(draft_cache_len)
            target_past.crop(target_cache_len)

            if tokenizer.eos_token_id is not None:
                if generated[0, -1].item() == tokenizer.eos_token_id:
                    break

    end_event.record()
    torch.cuda.synchronize()
    sd_time = start_event.elapsed_time(end_event) / 1000

    final_ids = generated[0, :target_len]
    sd_tokens = len(final_ids) - prompt_len
    sd_speed = sd_tokens / sd_time
    speedup = sd_speed / ar_speed
    acceptance_rate = accepted_tokens / total_draft_tokens if total_draft_tokens > 0 else 0
    average_gamma = total_draft_tokens / draft_steps if draft_steps > 0 else 0

    return sd_time, sd_speed, speedup, acceptance_rate, average_gamma, sd_tokens


# ==================================================
# Adaptive gamma strategies
# ==================================================

adaptive_results = {}

for strategy in ["heuristic", "confidence", "margin", "entropy"]:
    set_seed(42)
    result = run_sd(strategy)
    adaptive_results[strategy] = result

    print(
        f"{strategy}: {result[0]:.3f}s | "
        f"{result[1]:.3f} tok/s | "
        f"speedup={result[2]:.3f}x | "
        f"acceptance_rate={result[3]:.3f} | "
        f"average_gamma={result[4]:.3f} | "
        f"tokens={result[5]}"
    )


# ==================================================
# Best fixed gamma
# ==================================================

fixed_results = {}

for gamma in fixed_gammas:
    set_seed(42)
    result = run_sd("fixed", fixed_gamma=gamma)
    fixed_results[gamma] = result

    print(
        f"fixed gamma={gamma}: {result[0]:.3f}s | "
        f"{result[1]:.3f} tok/s | "
        f"speedup={result[2]:.3f}x | "
        f"acceptance_rate={result[3]:.3f} | "
        f"tokens={result[5]}"
    )

best_gamma = max(fixed_results, key=lambda gamma: fixed_results[gamma][1])
best_strategy = max(adaptive_results, key=lambda strategy: adaptive_results[strategy][1])

print(f"Best adaptive strategy: {best_strategy}")
print(f"Best fixed gamma: {best_gamma}")
print(
    f"Adaptive vs fixed: "
    f"{adaptive_results[best_strategy][1]:.3f} vs "
    f"{fixed_results[best_gamma][1]:.3f} tok/s"
)
```

```python
# ==================================================
# 异构词表复现：Union 与 SLEM
# ==================================================
#   target model: Qwen/Qwen2.5-1.5B-Instruct
#   draft model:  distilbert/distilgpt2

# Qwen2.5-1.5B
qwen_target_root = (
    Path.home()
    / ".cache"
    / "huggingface"
    / "hub"
    / "models--Qwen--Qwen2.5-1.5B-Instruct"
    / "snapshots"
)
qwen_target_path = next(
    path
    for path in qwen_target_root.iterdir()
    if (path / "config.json").exists()
)

target_tokenizer = AutoTokenizer.from_pretrained(
    qwen_target_path,
    local_files_only=True
)
target_model = AutoModelForCausalLM.from_pretrained(
    qwen_target_path,
    local_files_only=True,
    dtype=torch.float16 if device == "cuda" else torch.float32
).to(device)

draft_tokenizer = small_tokenizer
draft_model = small_model.to(
    device=device,
    dtype=torch.float16 if device == "cuda" else torch.float32
)

target_model.eval()
draft_model.eval()
target_model.generation_config.repetition_penalty = 1.0
draft_model.generation_config.repetition_penalty = 1.0

print("device:", device)
print("target tokenizer:", type(target_tokenizer).__name__)
print("draft tokenizer:", type(draft_tokenizer).__name__)
print("target vocabulary size:", target_model.config.vocab_size)
print("draft vocabulary size:", draft_model.config.vocab_size)
```

```python
# ==================================================
# Tokenizer 检查：词表交集、round-trip 与边界重分词
# ==================================================
# 展示异构 tokenizer 的几个典型问题：词表交集、文本归一化、边界切分。

# 此处我们将两Model词表的token全部解码成字符串，再一一对比
target_vocab_strings = set()
for token_id in range(len(target_tokenizer)):
    target_vocab_strings.add(target_tokenizer.decode([token_id]))

shared_count = 0
for token_id in range(len(draft_tokenizer)):
    token_text = draft_tokenizer.decode([token_id])
    if token_text in target_vocab_strings:
        shared_count += 1

print("shared token strings:", shared_count)
print("shared ratio in draft vocabulary:", shared_count / len(draft_tokenizer))

print("\nround-trip false examples:")

# 此处演示第一种情况，文本编码后再解码并不一定保持一致，此处例子为Codex寻找
texts = [
    "cafe\u0301",   # e + 组合重音符，看起来像 café，但是后者的是单独一个字符，这里是拼接起来的
    "\u212b",       # Angstrom sign，看起来像Å，Unicode码点不同。
    "\u1100\u1161", # 韩文分解形式，实际上是两个组合起来的，形式为가。
]

for text in texts:
    target_round_trip = target_tokenizer.decode(target_tokenizer.encode(text))
    draft_round_trip = draft_tokenizer.decode(draft_tokenizer.encode(text))

    print("\ntext:", repr(text))
    print("target round-trip:", repr(target_round_trip), "| preserved:", target_round_trip == text)
    print("draft round-trip:", repr(draft_round_trip), "| preserved:", draft_round_trip == text)


print("\nnon-injective examples:")
# 展示非单射问题，不同字符串可能被 tokenizer 编成同一组 token ID
pairs = [
    ("cafe\u0301", "café"),
    ("\u212b", "Å"),
    ("\u1100\u1161", "가"),
]

for left, right in pairs:
    target_left_ids = target_tokenizer.encode(left)
    target_right_ids = target_tokenizer.encode(right)
    draft_left_ids = draft_tokenizer.encode(left)
    draft_right_ids = draft_tokenizer.encode(right)
# target 和 draft 分别对以上例子编码
    print("\npair:", repr(left), repr(right))
    print("target same token ids:", target_left_ids == target_right_ids)
    print("draft same token ids:", draft_left_ids == draft_right_ids)

# 展示边界切分问题，分开tokenize后拼起来，不一定等于整体tokenize
print("\nboundary test:")

left_text = "hel"
right_text = "lo world"
full_text = left_text + right_text

target_separate = target_tokenizer.convert_ids_to_tokens(
    target_tokenizer.encode(left_text) + target_tokenizer.encode(right_text)
)
target_joint = target_tokenizer.convert_ids_to_tokens(target_tokenizer.encode(full_text))

draft_separate = draft_tokenizer.convert_ids_to_tokens(
    draft_tokenizer.encode(left_text) + draft_tokenizer.encode(right_text)
)
draft_joint = draft_tokenizer.convert_ids_to_tokens(draft_tokenizer.encode(full_text))

print("\ntarget tokenizer:")
print("separate tokens:", target_separate)
print("joint tokens:", target_joint)
print("same tokenization:", target_separate == target_joint)

print("\ndraft tokenizer:")
print("separate tokens:", draft_separate)
print("joint tokens:", draft_joint)
print("same tokenization:", draft_separate == draft_joint)


    


```

```python
# Union speculative decoding：异构 tokenizer 下先让 draft 提议，再让 target 验证。
prompt = "Once upon a time"
max_new_tokens = 64
gamma = 4

target_model.eval()
draft_model.eval()

target_inputs = target_tokenizer(prompt, return_tensors="pt").to(device) #用 target tokenizer 编码 prompt
target_ids = target_inputs.input_ids.clone() #当前已经确认的 target token 序列
start_length = target_ids.shape[1] #用来判断已经生成了多少新token

target_vocab_strings = {}
for token_id in range(len(target_tokenizer)):
    token_text = target_tokenizer.decode([token_id])
    if token_text not in target_vocab_strings:
        target_vocab_strings[token_text] = token_id
#维护一个字典，建立字符串到token_id的映射
draft_to_target = torch.full((len(draft_tokenizer),), -1, dtype=torch.long, device=device)
for token_id in range(len(draft_tokenizer)):
    token_text = draft_tokenizer.decode([token_id])
    if token_text in target_vocab_strings:
        draft_to_target[token_id] = target_vocab_strings[token_text]
#这段遍历 draft tokenizer 的每个 token。如果某个 draft token 解码后的字符串也存在于 target tokenizer 的单 token 字符串里，就记录映射关系，不存在记为-1
accepted_count = 0 #被 target 接受的 draft token 数
checked_count = 0 #被 target 检查过的 draft token 数
outside_count = 0 #draft token 不在 target 单 token 词表里的次数

print("prompt:", repr(prompt))
print("max_new_tokens:", max_new_tokens)
print("gamma:", gamma)

with torch.no_grad():
    while target_ids.shape[1] < start_length + max_new_tokens:
        current_text = target_tokenizer.decode(target_ids[0], skip_special_tokens=True)
        draft_inputs = draft_tokenizer(current_text, return_tensors="pt").to(device)

        draft_output = draft_model.generate(
            **draft_inputs,
            max_new_tokens=gamma,
            do_sample=True,
            temperature=1.0,
            output_scores=True,
            return_dict_in_generate=True,
            pad_token_id=draft_tokenizer.eos_token_id,
        )

        draft_new_ids = draft_output.sequences[0, draft_inputs.input_ids.shape[1]:] # 切掉 prompt

        mapped_target_ids = [] #draft token 对应的 target token ID
        draft_probs = [] #draft 给该 token 的概率 q
        draft_texts = [] #draft token 的字符串形式，方便打印

        for step in range(len(draft_new_ids)):
            draft_token_int = int(draft_new_ids[step].item())
            target_token_int = int(draft_to_target[draft_token_int].item())
            draft_text = draft_tokenizer.decode([draft_token_int])

            if target_token_int < 0:
                outside_count += 1
                print("\ndraft token outside target vocabulary:", repr(draft_text))
                break

            step_probs = torch.softmax(draft_output.scores[step][0].float(), dim=-1)

            mapped_target_ids.append(target_token_int)
            draft_probs.append(float(step_probs[draft_token_int].item()))
            draft_texts.append(draft_text)

        if len(mapped_target_ids) == 0:
            target_output = target_model(target_ids)
            target_probs = torch.softmax(target_output.logits[:, -1, :].float(), dim=-1)[0]
            target_probs[len(target_tokenizer):] = 0
            target_probs = target_probs / target_probs.sum()

            next_id = torch.multinomial(target_probs, num_samples=1).view(1, 1)
            target_ids = torch.cat([target_ids, next_id], dim=1)
            continue

        candidate_ids = torch.tensor([mapped_target_ids], dtype=torch.long, device=device)
        verify_ids = torch.cat([target_ids, candidate_ids], dim=1)

        target_output = target_model(verify_ids)
        rejected = False

        print("\ndraft proposes:", [repr(text) for text in draft_texts])

        for step in range(len(mapped_target_ids)):
            checked_count += 1

            target_token_int = mapped_target_ids[step]
            position = target_ids.shape[1] + step - 1

            current_target_probs = torch.softmax(
                target_output.logits[:, position, :].float(),
                dim=-1,
            )[0]

            # Qwen 的模型 vocab 可能大于 tokenizer 长度，采样前把 tokenizer 外的预留位置清掉。
            current_target_probs[len(target_tokenizer):] = 0
            current_target_probs = current_target_probs / current_target_probs.sum()

            q = draft_probs[step]
            p = float(current_target_probs[target_token_int].item())
            accept_prob = min(1.0, p / q) if q > 0 else 1.0
            random_value = float(torch.rand(1, device=device).item())

            print(
                "check:",
                repr(draft_texts[step]),
                "| p:",
                round(p, 4),
                "| q:",
                round(q, 4),
                "| accept:",
                round(accept_prob, 4),
            )

            if random_value <= accept_prob:
                accepted_count += 1
                accepted_id = torch.tensor([[target_token_int]], dtype=torch.long, device=device)
                target_ids = torch.cat([target_ids, accepted_id], dim=1)
            else:
                draft_probs_on_target = torch.zeros(
                    current_target_probs.shape[0],
                    dtype=torch.float32,
                    device=device,
                )

                valid_mask = draft_to_target >= 0
                valid_target_ids = draft_to_target[valid_mask]

                current_draft_probs = torch.softmax(draft_output.scores[step][0].float(), dim=-1)
                draft_probs_on_target.index_add_(
                    0,
                    valid_target_ids,
                    current_draft_probs[valid_mask],
                )

                residual_probs = torch.clamp(current_target_probs - draft_probs_on_target, min=0)
                residual_probs[len(target_tokenizer):] = 0
                residual_sum = residual_probs.sum()

                if residual_sum <= 0:
                    residual_probs = current_target_probs
                else:
                    residual_probs = residual_probs / residual_sum

                next_id = torch.multinomial(residual_probs, num_samples=1).view(1, 1)
                target_ids = torch.cat([target_ids, next_id], dim=1)
                rejected = True

                print("rejected, target samples:", repr(target_tokenizer.decode(next_id[0])))
                break

        if not rejected:
            next_position = target_ids.shape[1] - 1
            target_probs = torch.softmax(target_output.logits[:, next_position, :].float(), dim=-1)[0]
            target_probs[len(target_tokenizer):] = 0
            target_probs = target_probs / target_probs.sum()

            next_id = torch.multinomial(target_probs, num_samples=1).view(1, 1)
            target_ids = torch.cat([target_ids, next_id], dim=1)

            print("all accepted, target samples extra token:", repr(target_tokenizer.decode(next_id[0])))

        if target_ids[0, -1].item() == target_tokenizer.eos_token_id:
            break

target_text = target_tokenizer.decode(target_ids[0], skip_special_tokens=True)

print("\nUnion result:")
print(target_text)
print("\naccepted draft tokens:", accepted_count)
print("checked draft tokens:", checked_count)
print("outside target vocabulary:", outside_count)
print("acceptance rate:", accepted_count / checked_count if checked_count > 0 else 0.0)
```

```python
# SLEM / UAD 思路演示：draft 先生成文本，再让 target tokenizer 重新编码并验证。
prompt = "Once upon a time"
max_new_tokens = 64
gamma = 4
lookbehind = 20

target_model.eval()
draft_model.eval()

target_inputs = target_tokenizer(prompt, return_tensors="pt").to(device)
target_ids = target_inputs.input_ids.clone()
start_length = target_ids.shape[1]

accepted_count = 0
checked_count = 0

print("prompt:", repr(prompt))
print("max_new_tokens:", max_new_tokens)
print("gamma:", gamma)
print("lookbehind:", lookbehind)

with torch.no_grad():
    while target_ids.shape[1] < start_length + max_new_tokens:
        current_text = target_tokenizer.decode(
            target_ids[0],
            skip_special_tokens=True,
            clean_up_tokenization_spaces=False,
        )

        draft_inputs = draft_tokenizer(
            current_text,
            return_tensors="pt",
            add_special_tokens=False,
        ).to(device)

        draft_output = draft_model.generate(
            **draft_inputs,
            max_new_tokens=gamma,
            do_sample=False,
            return_dict_in_generate=True,
            pad_token_id=draft_tokenizer.eos_token_id,
        )

        draft_full_text = draft_tokenizer.decode(
            draft_output.sequences[0],
            skip_special_tokens=True,
            clean_up_tokenization_spaces=False,
        )

        # SLEM 的关键：不直接映射 draft token，而是把 draft 生成后的完整文本交给 target tokenizer 重编码。
        reencoded_ids = target_tokenizer(
            draft_full_text,
            return_tensors="pt",
            add_special_tokens=False,
        ).input_ids.to(device)

        old_tokens = target_ids[0, -lookbehind:].tolist()
        new_tokens = reencoded_ids[0].tolist()

        best_old_start = 0
        best_new_start = 0
        best_length = 0

        # 在 target 当前后缀和 target 重编码结果之间找最长公共片段。
        # 公共片段之后的 token，就是 draft 通过文本提出的 target 候选 token。
        for old_start in range(len(old_tokens)):
            for new_start in range(len(new_tokens)):
                length = 0

                while (
                    old_start + length < len(old_tokens)
                    and new_start + length < len(new_tokens)
                    and old_tokens[old_start + length] == new_tokens[new_start + length]
                ):
                    length += 1
#old_tokens 和 new_tokens 之间最长的一段连续相同 token
                old_end = old_start + length
                best_old_end = best_old_start + best_length

                if length > best_length or (
                    length == best_length and old_end > best_old_end
                ):
                    best_old_start = old_start
                    best_new_start = new_start
                    best_length = length
#长度相同时，优先靠近旧序列末尾
        candidate_start = best_new_start + best_length
        candidate_ids = reencoded_ids[:, candidate_start:]

        if candidate_ids.shape[1] == 0:
            target_output = target_model(target_ids)
            next_id = target_output.logits[:, -1, :].argmax(dim=-1, keepdim=True)
            target_ids = torch.cat([target_ids, next_id], dim=1)
            print("\nno aligned candidate, target adds:", repr(target_tokenizer.decode(next_id[0])))
            continue

        candidate_ids = candidate_ids[:, :gamma]
        verify_ids = torch.cat([target_ids, candidate_ids], dim=1)
        target_output = target_model(verify_ids)

        print("\ndraft text tail:", repr(draft_full_text[-80:]))
        print("aligned candidate:", repr(target_tokenizer.decode(candidate_ids[0])))

        rejected = False

        for step in range(candidate_ids.shape[1]):
            checked_count += 1

            position = target_ids.shape[1] + step - 1
            expected_id = candidate_ids[0, step].item()
            target_next_id = target_output.logits[:, position, :].argmax(dim=-1).item()

            expected_text = target_tokenizer.decode([expected_id])
            target_text = target_tokenizer.decode([target_next_id])

            print(
                "check:",
                repr(expected_text),
                "| target greedy:",
                repr(target_text),
                "| accepted:",
                expected_id == target_next_id,
            )

            if expected_id == target_next_id:
                accepted_count += 1
                target_ids = torch.cat(
                    [
                        target_ids,
                        torch.tensor([[expected_id]], dtype=torch.long, device=device),
                    ],
                    dim=1,
                )
            else:
                target_ids = torch.cat(
                    [
                        target_ids,
                        torch.tensor([[target_next_id]], dtype=torch.long, device=device),
                    ],
                    dim=1,
                )
                rejected = True
                break

        if not rejected:
            next_position = target_ids.shape[1] - 1
            next_id = target_output.logits[:, next_position, :].argmax(dim=-1, keepdim=True)
            target_ids = torch.cat([target_ids, next_id], dim=1)
            print("all accepted, target adds:", repr(target_tokenizer.decode(next_id[0])))

        if target_ids[0, -1].item() == target_tokenizer.eos_token_id:
            break

result_text = target_tokenizer.decode(
    target_ids[0],
    skip_special_tokens=True,
    clean_up_tokenization_spaces=False,
)

print("\nSLEM result:")
print(result_text)
print("\naccepted candidate tokens:", accepted_count)
print("checked candidate tokens:", checked_count)
print("acceptance rate:", accepted_count / checked_count if checked_count > 0 else 0.0)
```

```python
# TLI 演示：只在 target 和 draft 共享 token 集合内采样，减少异构 tokenizer 的无效候选。
prompt = "Once upon a time"
max_new_tokens = 64
gamma = 4

target_model.eval()
draft_model.eval()

target_vocab_strings = {}
for token_id in range(len(target_tokenizer)):
    token_text = target_tokenizer.decode([token_id])
    if token_text not in target_vocab_strings:
        target_vocab_strings[token_text] = token_id
#遍历 target tokenizer 的所有 token ID，维护一个字典存储解码的字符串
draft_to_target = torch.full((len(draft_tokenizer),), -1, dtype=torch.long, device=device)
for token_id in range(len(draft_tokenizer)):
    token_text = draft_tokenizer.decode([token_id])
    if token_text in target_vocab_strings:
        draft_to_target[token_id] = target_vocab_strings[token_text]
#遍历draft tokenizer 的所有 token，如果某个draft token 解码出来的字符串，也存在于target tokenizer 的单 token 字符串集合中，就建立映射，其实和Union就差不多
valid_draft_mask = draft_to_target >= 0 #True  -> 这个 draft token 在 target 词表中有对应单 token，反之则否

target_inputs = target_tokenizer(prompt, return_tensors="pt").to(device)
target_ids = target_inputs.input_ids.clone()
start_length = target_ids.shape[1]

accepted_count = 0
checked_count = 0

print("prompt:", repr(prompt))
print("max_new_tokens:", max_new_tokens)
print("gamma:", gamma)
print("valid draft token ratio:", round(valid_draft_mask.float().mean().item(), 4))

with torch.no_grad():
    while target_ids.shape[1] < start_length + max_new_tokens:
        current_text = target_tokenizer.decode(
            target_ids[0],
            skip_special_tokens=True,
            clean_up_tokenization_spaces=False,
        )
        draft_inputs = draft_tokenizer(
            current_text,
            return_tensors="pt",
            add_special_tokens=False,
        ).to(device)

        draft_ids = []
        draft_texts = []

        for _ in range(gamma):
            draft_output = draft_model(**draft_inputs)
            draft_logits = draft_output.logits[:, -1, :].float() # 让 draft 模型输出下一个 token 的 logits

            draft_logits[:, ~valid_draft_mask] = -float("inf") # 把交集外 token 的 logits 全部设为inf，~符号是bool取反操作，这样执行argmax就不会选到
            draft_next_id = draft_logits.argmax(dim=-1, keepdim=True)

            draft_ids.append(int(draft_next_id.item()))
            draft_texts.append(draft_tokenizer.decode(draft_next_id[0]))

            draft_inputs.input_ids = torch.cat([draft_inputs.input_ids, draft_next_id], dim=1)
            draft_inputs.attention_mask = torch.cat(
                [
                    draft_inputs.attention_mask,
                    torch.ones((1, 1), dtype=draft_inputs.attention_mask.dtype, device=device),
                ],
                dim=1,
            )
            #把新生成的draft token 拼回draft 输入
        candidate_ids = torch.tensor(
            [[int(draft_to_target[draft_id].item()) for draft_id in draft_ids]],
            dtype=torch.long,
            device=device,
        )
        # draft token 映射成 target token

        verify_ids = torch.cat([target_ids, candidate_ids], dim=1)
        target_output = target_model(verify_ids)

        print("\nTLI draft:", [repr(text) for text in draft_texts])
        print("mapped target candidate:", repr(target_tokenizer.decode(candidate_ids[0])))

        rejected = False

        for step in range(candidate_ids.shape[1]):
            checked_count += 1

            position = target_ids.shape[1] + step - 1
            expected_id = int(candidate_ids[0, step].item())
            target_next_id = int(target_output.logits[:, position, :].argmax(dim=-1).item())

            expected_text = target_tokenizer.decode([expected_id])
            target_text = target_tokenizer.decode([target_next_id])

            accepted = expected_id == target_next_id

            print(
                "check:",
                repr(expected_text),
                "| target greedy:",
                repr(target_text),
                "| accepted:",
                accepted,
            )

            if accepted:
                accepted_count += 1
                target_ids = torch.cat(
                    [target_ids, torch.tensor([[expected_id]], dtype=torch.long, device=device)],
                    dim=1,
                )
            else:
                target_ids = torch.cat(
                    [target_ids, torch.tensor([[target_next_id]], dtype=torch.long, device=device)],
                    dim=1,
                )
                rejected = True
                break
            #拒绝时用target token 替换

        if not rejected:
            next_position = target_ids.shape[1] - 1
            next_id = target_output.logits[:, next_position, :].argmax(dim=-1, keepdim=True)
            target_ids = torch.cat([target_ids, next_id], dim=1)
            print("all accepted, target adds:", repr(target_tokenizer.decode(next_id[0])))

        if target_ids[0, -1].item() == target_tokenizer.eos_token_id:
            break

result_text = target_tokenizer.decode(
    target_ids[0],
    skip_special_tokens=True,
    clean_up_tokenization_spaces=False,
)

print("\nTLI result:")
print(result_text)
print("\naccepted candidate tokens:", accepted_count)
print("checked candidate tokens:", checked_count)
print("acceptance rate:", accepted_count / checked_count if checked_count > 0 else 0.0)
```

```python
# Speculative Sampling based on Tree Sampling, doesn't implement KV_CACHE properly
# Took References: https://github.com/bassrehab/speculative-decoding/blob/main/tree_speculation.py 


import torch
import torch.nn.functional as F
import time

from transformers import AutoModelForCausalLM, AutoTokenizer, PreTrainedModel
from dataclasses import dataclass, field 

@dataclass
class TreeNode:
    ''' Represents a node in the speculation tree'''
    token_id: int
    prob: float
    depth: int
    parent: Optional['TreeNode'] = None
    children: List['TreeNode'] = field(default_factory=list)
    position_in_sequence: int = 0

    def is_leaf(self):
        ''' Returns if current node is a leaf node'''
        return len(self.children) == 0
    
    def get_path_to_root(self):
        '''Get token IDs from root to current node'''
        path = []
        node = self
        while node:
            path.append(node.token_id)
            node = node.parent
        
        return list(reversed(path))

@dataclass
class TreeSpecConfig:
    '''Tree Speculation Configuration'''
    draft_model_name: str = "gpt2" # Draft model
    target_model_name: str = "gpt2-large" # Target model
    max_depth: int = 4 # gamma
    branch_factor: int = 2 # Guess top 2 most likely words
    max_candidates: int = 16 # Absolute maximum number of guesses allowed
    temperature: float = 0.9 # How random the AI is
    top_p: float = 0.9  # Adjust the choice based on probability 
    max_new_tokens: int = 100 # Max number of tokens to guess


@dataclass
class TreeMetrics:
    '''Tree speculation matrics'''
    total_tokens_generated: int = 0 
    total_candidates_evaluated: int = 0
    total_iterations: int = 0 # How many times target had to review a batch
    total_time_ms: float = 0.0
    tree_depths: List[int] = field(default_factory=list) # How many words target accepted on each turn


    @property
    def avg_accepted_per_iteration(self):
        return self.total_tokens_generated / max(self.total_iterations, 1) # To prevent division by 0 errors
    
    @property
    def tokens_per_second(self):
        return (self.total_tokens_generated / max(self.total_time_ms, 1)) * 1000 
    
    @property
    def avg_tree_depth(self):
        return sum(self.tree_depths) / max(len(self.tree_depths), 1)

class SpeculationTree:
    def __init__(self, config):
        self.config = config
        self.root = None
        self.nodes = []
        self.leaves = []
    
    def build_tree(self, draft_model, input_ids, past_key_values, device):
        '''
        Build speculation tree using draft model.
        Returns tuple of candidate tokens, candidate probs, and tree attention mask
        '''
        self.nodes = []
        self.leaves = []

        # Root node, last token in sequence
        self.root = TreeNode(
            token_id = input_ids[0,-1].item(),
            prob = 1.0,
            depth = 0,
            position_in_sequence = 0
        )

        self.nodes.append(self.root)

        # BFS
        current_level = [self.root]
        position = 1

        prefix_ids = input_ids[0, :-1].tolist()

        for depth in range(1, self.config.max_depth + 1):
            if len(self.nodes) >= self.config.max_candidates:
                break
            
            next_level = []
            
            for parent in current_level:
                if len(self.nodes) >= self.config.max_candidates:
                    break
                
                path_tokens = parent.get_path_to_root()
                full_sequence = prefix_ids + path_tokens
                path_tensor = torch.tensor([full_sequence], device=device)

                with torch.no_grad():
                    outputs = draft_model(path_tensor, use_cache=False)
                    logits = outputs.logits[0, -1, :]
                
                probs = F.softmax(logits / max(self.config.temperature, 1e-8), dim = -1)
                top_probs, top_indices = torch.topk(probs, self.config.branch_factor)

                # Creating child nodes
                for i in range(self.config.branch_factor):
                    if len(self.nodes) >= self.config.max_candidates:
                        break
                    
                    child = TreeNode(
                        token_id = top_indices[i].item(),
                        prob = top_probs[i].item(),
                        depth = depth,
                        parent = parent,
                        position_in_sequence = position
                    )

                    parent.children.append(child)
                    self.nodes.append(child)
                    next_level.append(child)
                    position += 1
            
            current_level = next_level
            if not current_level:
                break
            
            # Leaves
            self.leaves = [n for n in self.nodes if n.is_leaf()]

            # Flatten tree 
        return self.flatten_tree(device)           
    
    def flatten_tree(self, device):
        '''Flatten tree into tensors for batched verification, returns tokens, probs, attention_mask'''
        num_nodes = len(self.nodes)

        # Token ID of each node
        tokens = torch.tensor(
            [n.token_id for n in self.nodes],
            dtype = torch.long,
            device=device
        )

        # Probability of each node
        probs = torch.tensor(
            [n.prob for n in self.nodes],
            dtype = torch.float,
            device=device
        )

        # Attention mask, so that each node can only attend to tokens before itself
        attention_mask = torch.zeros(num_nodes, num_nodes, device=device)

        for i, node in enumerate(self.nodes):
            # Can attend itself
            attention_mask[i, i] = 1

            ancestor = node.parent
            while ancestor is not None:
                ancestor_idx = self.nodes.index(ancestor)
                attention_mask[i, ancestor_idx] = 1
                ancestor = ancestor.parent
            
        return tokens, probs, attention_mask

    def get_paths(self):
        """Returns a list of paths where each path is a lost of nodes from root to leaf"""
        paths = []
        for leaf in self.leaves:
            path = []
            node = leaf
            while node:
                path.append(node)
                node = node.parent
            
            paths.append(list(reversed(path)))
        
        return paths

@torch.no_grad()
def verify_tree(
    target_model, 
    input_ids, 
    tree, 
    candidate_tokens, 
    candidate_probs, 
    tree_attention_mask,
    past_key_values,
    temperature=1.0,
    device="cuda"):
    '''
    Verifies all tree paths with target model and select best accepted sequence
    Returns best path and best length
    '''
    paths = tree.get_paths()
    if not paths:
        return [], 0
        
    prefix_ids = input_ids[0, :-1].tolist()
    
    # Batch all valid root-to-leaf paths
    batch_input_ids = []
    for path in paths:
        path_tokens = [node.token_id for node in path]
        batch_input_ids.append(prefix_ids + path_tokens)
        
    # Pad sequences to max length in this batch for the standard forward pass
    max_len = max(len(seq) for seq in batch_input_ids)
    padded_batch = []
    attention_mask = []
    
    for seq in batch_input_ids:
        pad_len = max_len - len(seq)
        padded_batch.append(seq + [0] * pad_len) 
        attention_mask.append([1] * len(seq) + [0] * pad_len)
        
    batch_tensor = torch.tensor(padded_batch, device=device)
    mask_tensor = torch.tensor(attention_mask, device=device)
    
    outputs = target_model(batch_tensor, attention_mask=mask_tensor, use_cache=False)
    batch_logits = outputs.logits 
    
    best_path = []
    best_length = 0
    seq_len_prefix = len(prefix_ids)
    
    for path_idx, path in enumerate(paths):
        accepted = []
        path_logits = batch_logits[path_idx]
        
        for i, node in enumerate(path[1:], start=1):
            logit_idx = seq_len_prefix + i - 1
            
            target_probs_step = F.softmax(path_logits[logit_idx] / max(temperature, 1e-8), dim=-1)
            p_target = target_probs_step[node.token_id].item()
            p_draft = node.prob
            
            accept_prob = min(1.0, p_target / (p_draft + 1e-10))
            
            # Coin Toss and Adjusted Probability Distribution
            if torch.rand(1).item() < accept_prob:
                accepted.append(node.token_id)
            else:
                adjusted = target_probs_step.clone()
                adjusted[node.token_id] = max(0, adjusted[node.token_id] - p_draft)
                adjusted = adjusted / (adjusted.sum() + 1e-10)
                
                # Normalization
                if adjusted.sum() > 1e-8:
                    new_token = torch.multinomial(adjusted, 1).item()
                else:
                    new_token = path_logits[logit_idx].argmax().item()
                
                accepted.append(new_token)
                break 
        
        # Update best path
        if len(accepted) > best_length:
            best_length = len(accepted)
            best_path = accepted
            
            # Bonus token if the entire branch was accepted
            if len(accepted) == len(path) - 1:
                last_logit_idx = seq_len_prefix + len(path) - 1
                bonus_probs = F.softmax(path_logits[last_logit_idx] / max(temperature, 1e-8), dim=-1)
                bonus_token = torch.multinomial(bonus_probs, 1).item()
                best_path.append(bonus_token)
                best_length += 1
                
    return best_path, best_length

@torch.no_grad()
def tree_speculative_decode(draft_model, target_model, tokenizer, prompt, config, device):
    '''
    Generate text using tree-based speculative decoding, returns generated text and metrics
    '''
    # Metrics object, gets updated in each iteration
    metrics = TreeMetrics()
    start_time = time.perf_counter()

    input_ids = tokenizer.encode(prompt, return_tensors="pt").to(device)
    all_token_ids = input_ids.clone()

    generated_tokens = 0
    tree = SpeculationTree(config)

    while generated_tokens < config.max_new_tokens:
        metrics.total_iterations += 1

        # Speculation Tree
        candidate_tokens, candidate_probs, tree_mask = tree.build_tree(
            draft_model, all_token_ids, None, device
        )

        metrics.total_candidates_evaluated += len(candidate_tokens)

        # Verify with target
        accepted_tokens, num_accepted = verify_tree(
            target_model, all_token_ids, tree,
            candidate_tokens, candidate_probs, tree_mask,
            None, config.temperature, device
        )

        if not accepted_tokens:
            with torch.no_grad():
                outputs = target_model(all_token_ids, use_cache=False)
                logits = outputs.logits[0, -1, :]
                probs = F.softmax(logits / config.temperature, dim=-1)
                token = torch.multinomial(probs, 1).item()
                accepted_tokens = [token]
            
        
        # Update sequence
        new_tokens = torch.tensor([accepted_tokens], device=device)
        all_token_ids = torch.cat([all_token_ids, new_tokens], dim=-1)
        generated_tokens += len(accepted_tokens)

        metrics.total_tokens_generated += len(accepted_tokens)
        metrics.tree_depths.append(len(accepted_tokens))

        if tokenizer.eos_token_id in accepted_tokens:
            break
        
        metrics.total_time_ms = (time.perf_counter() - start_time) * 1000
        generated_text = tokenizer.decode(all_token_ids[0], skip_special_tokens=True)

    return generated_text, metrics


def normal_inference(model, tokenizer, prompt, max_new_tokens=30):
    inputs = tokenizer(prompt, return_tensors='pt').to(device)
    outputs = model.generate(inputs['input_ids'], max_new_tokens=max_new_tokens, attention_mask=inputs["attention_mask"],  pad_token_id=tokenizer.eos_token_id, return_dict_in_generate=True, output_scores=True, do_sample=True, temperature=0.9, top_p=0.9)
    text = tokenizer.decode(outputs.sequences[0], skip_special_tokens=True)
    input_len = inputs["input_ids"].shape[1]
    new_tokens = outputs.sequences[0][input_len:]

    distributions = []
    token_probs = []

    for token_id, score in zip(new_tokens, outputs.scores):
        prob_dist = torch.softmax(score[0], dim=-1)
        prob = prob_dist[token_id].item()

        distributions.append(prob_dist)
        token_probs.append((tokenizer.decode([token_id]), prob))

    return text, new_tokens, token_probs, distributions

config = TreeSpecConfig()
device = "cuda" if torch.cuda.is_available() else "cpu"
draft_model = AutoModelForCausalLM.from_pretrained(config.draft_model_name).to(device).eval()

tokenizer = AutoTokenizer.from_pretrained(config.target_model_name)
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token


target_model = AutoModelForCausalLM.from_pretrained(config.target_model_name).to(device).eval()
prompt = "Once upon a time there was a little girl named Alice "

print(f"Prompt: {prompt}")
print(f"Using device: {device}")
print(f"Draft model: {config.draft_model_name}")
print(f"Target model: {config.target_model_name}")
print(f"Tree depth: {config.max_depth}")
print(f"Branch factor: {config.branch_factor}")
print(f"Max candidates: {config.max_candidates}")
print("-" * 60)


start_normal = time.perf_counter()
normal = normal_inference(target_model, tokenizer, prompt, 100)
end_normal = time.perf_counter()

print("Normal latency: ", (end_normal - start_normal) * 1000)
print("Normal output: ", normal[0])
print("-" * 60)
print("\n")


text, metrics = tree_speculative_decode(
    draft_model, target_model, tokenizer, prompt, config, device
)

print("\n" + "=" * 60)
print("TREE SPECULATION RESULTS")
print("=" * 60)
print(f"Tokens generated: {metrics.total_tokens_generated}")
print(f"Total iterations: {metrics.total_iterations}")
print(f"Avg tokens/iteration: {metrics.avg_accepted_per_iteration:.2f}")
print(f"Avg tree depth accepted: {metrics.avg_tree_depth:.2f}")
print(f"Candidates evaluated: {metrics.total_candidates_evaluated}")
print(f"Time: {metrics.total_time_ms:.1f} ms")
print(f"Throughput: {metrics.tokens_per_second:.1f} tok/s")
print(f"Speculative output: ", text)
```
